```python
hr_sal = df[df["dept"]=="HR"]["Salray"]
hr_sal
```

```python
ca_sal = df[df["dept"]=="CA"]["Salray"]
ca_sal
```

```python
#boxplot
```

```python
plt.boxplot([hr_sal , it_sal , ca_sal , fi_sal] , labels=["HR" , "IT", "CA","Fincance"])
plt.legend()
plt.show()
```

```python
#Pie chart
```

```python
salray_by_dept=df.groupby("dept")["Salray"].sum()
salray_by_dept
```

```python
plt.pie(salray_by_dept , labels=["CA","Finnance","HR","IT"], autopct = "%1.1f" , shadow=True)
plt.title("Dept vs Salary")
plt.axis("equal")
plt.grid()
plt.show()
```

```python
#bar plot
```

```python
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fi_mean = sum(fi_sal)/len(fi_sal)
ca_mean = sum(ca_sal)/len(ca_sal)

```

```python
plt.bar(["Hr","It","Finance","Ca"],[hr_mean , it_mean , fi_mean , ca_mean] , color =["Red","Black","Red","Black"])
plt.title("Mean salray of each dept")
plt.grid()
plt.show()
```

```python
plt.scatter(df["Age"], df["Salray"] , c = df["dept"].map({"HR":"Yellow" , "Finance":"Orange" , "CA":"Red" , "IT":"Brown"}))
plt.xlabel("Age")
plt.ylabel("Salray")
plt
plt.show()
```

```python
color = {"HR":"Yellow" , "Finance":"Orange" , "CA":"Red" , "IT":"Brown"}
for dept, color in color.items():
    df_dept= df[df["dept"]==dept]
    plt.scatter(df_dept["Age"] , df_dept["Salray"] , c = color , label = dept)
    plt.legend()
    plt.xlabel("Age")
    plt.ylabel("Salray")
    plt.title("Dept wise salray")

```
