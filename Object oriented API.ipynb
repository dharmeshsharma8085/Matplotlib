```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig, axs = plt.subplots(1,3, figsize = (10,5))
```

```python
# universal _ Numeical
data = {
    "Salray":[25000,40000,24434,242453,12445,6544,123454,87600,56000,34000,45000,65400,34576,12098,45565,56000,34000,76000,56987,34876]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df["dept"]= ["HR" , "IT" ,"Finance","CA"]*5
```

```python
df["Age"] = [34,23,34,24,36,27,25,37,47,25,38,43,26,37,24,39,34,26,21,20]
df
```

```python
df["Experience"]=[1,2,3,1.5,3.3]*4
df.head()
```

```python
sort_age=df.sort_values("Age")
```

```python
# Combine all
axsfig, axs = plt.subplots(1,3, figsize = (10,5))

#line plot
axs[0].plot(sort_age["Age"],df["Salray"],color="blue" , marker ="*" , linewidth = "2")
axs[0].grid()
axs[0].set_title("Line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salray")

#HISTOGRAM

axs[1].hist(df["Salray"] , bins = 5 , color ="skyblue")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salray")
axs[1].set_ylabel("Frequency")

#boxplot
axs[2].boxplot(df["Salray"])
axs[2].set_title("Boxplot")
axs[2].set_xlabel("Salray")

plt.savefig("Objectplot API.png")
plt.show()


```

```python
data = {
    "Year" : [2020 , 2021 , 2022 , 2023],
    "Sales" : [100,150,200,250],
    "Profit" : [20,30,40,50],
    "Expense" : [80,120,160,200]
}
df1 = pd.DataFrame(data)
df1
```

```python
plt.plot(df1["Year"] , df1["Sales"] , label ="Sales")
plt.plot(df1["Year"] , df1["Profit"] , label ="Profit")
plt.plot(df1["Year"] , df1["Expense"] , label ="Expense")
plt.title("Financil Graph of company")
plt.xlabel("Years")
plt.ylabel("Amount")
plt.legend()
plt.show()
```
